```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('dataset/made.csv')
df
```

```python
df.shape
```

```python
df.describe()
```

```python
labels = np.array(df['infected'])
labels[:5]
```

```python
feature = df.drop(['infected','age'], axis=1)
# feature = df.drop('age', axis=1)
print(feature[:10])
feature = np.array(feature)
feature[:5]
```

```python
rf = RandomForestRegressor(n_estimators= 100, random_state = 42)
```

```python
X_train, X_test, y_train, y_test = train_test_split(feature, labels, test_size=0.2, random_state=10)
```

```python
print("Training data Size: ", X_train.shape)
print("Test data Sie: ", X_test.shape)
```

```python
rf.fit(X_train, y_train)
```

```python
prediction = rf.predict(X_test)
prediction[:10]
```

```python
print("Actual Value: ", y_test[:5])
print("Predicted Value: ", prediction[:5])
```
